#### Este notebook efetua o download dos arquivos 7z multipart e os extrai em uma pasta temporaria como arquivo completo.

py7zr - library que contem a funcao SevenZipFile(), utilizada para extrair o arquivo compactado.  
multivolumefile - library utilizada para identificar o arquivo multipart e juntar em uma unica parte para ser extraido.  
wget - funcao utilizada para efetuar download de arquivos.

```
import py7zr
import multivolumefile
```

```
def unzip_7z_file(file_name, in_path, out_path):
  """
    @Description: Unpack the multipart 7zip file and save the unpacked file in a temporary folder.
    @Param file_name: A string value containing the given file name.
    @Param in_path: A string value containing the given file origin path.
    @Param out_path: A string value containing the given file destination path.
  """

  try:
    with multivolumefile.open(f'{in_path}/{file_name}', mode='rb') as target_archive:
        with py7zr.SevenZipFile(target_archive, 'r') as archive:
            archive.extractall(path=f'{out_path}')
  except IndexError as e:
    print(f'No file was found! {e}')            
```
